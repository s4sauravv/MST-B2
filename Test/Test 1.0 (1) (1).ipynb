```python
def find_pairs(nums, target_sum):
    pairs =[]
    checked = set()

    for num in nums:
        complement = target_sum - num

        if complement in checked:
            pairs.append([complement, num])
        checked.add(num)

    return pairs

nums = [4, 5, 6, 9, 8, 6, 3]
target_sum = 9
result = find_pairs(nums, target_sum)
print(result)
```
